```python
def convert_currency(val):
    """
    Преобразует числовое значение строки в число с плавающей точкой:
     - удаляет пробелы ' '
     - преобразует число 0 в NaN
     - преобразует в число с плавающей точкой
    """
    if isinstance(val, str): 
        val = val.replace(' ', '').replace('O', '')
    if val==0: 
         val = np.nan
    return float(val)
```

```python
for i in range(1,10):
    st='2016-0'+str(i)+'-01'
    df_data_mini_16[st]=df_data_mini_16[st].apply(convert_currency)
df_data_mini_16['2016-11-01']=df_data_mini_16['2016-11-01'].apply(convert_currency)
df_data_mini_16['2016-12-01']=df_data_mini_16['2016-12-01'].apply(convert_currency)
```
